### Steps to Process ACS 2014 1-Year Median Contract Rent Data

1. **Load the Data**
   - Download the ACS 2014 1-year median contract rent data from the Census Bureau's American FactFinder website.

2. **Parse Principal City/State**
   - Extract the principal city and state from each metropolitan and micropolitan area name.

3. **Geocode City/State**
   - Convert the principal city and state information to latitude-longitude coordinates.

4. **Save to CSV**
   - Save the processed data into a CSV file.


```python
# Import the necessary libraries
import pandas as pd  # pandas for data manipulation and analysis
import time  # time for time-related functions (e.g., delays)
import requests  # requests for making HTTP requests to geocode API
import json  # json for handling JSON data

```

```python
# Load the ACS rent data from a CSV file into a DataFrame
# The CSV file contains ACS 2014 1-year median contract rent data
df = pd.read_csv('data/ACS_14_1YR_B25058_metro_micro/ACS_14_1YR_B25058.csv')

# Rename the columns to more intuitive names
# 'GEO.id' -> 'geo_id': Unique identifier for the geographical area
# 'GEO.id2' -> 'geo_id2': Another unique identifier for the geographical area
# 'GEO.display-label' -> 'msa_name': Name of the Metropolitan Statistical Area (MSA) or Micropolitan Statistical Area
# 'HD01_VD01' -> 'median_contract_rent': Median contract rent value
# 'HD02_VD01' -> 'margin_error': Margin of error for the median contract rent value
df = df.rename(columns={'GEO.id': 'geo_id', 
                        'GEO.id2': 'geo_id2', 
                        'GEO.display-label': 'msa_name', 
                        'HD01_VD01': 'median_contract_rent',
                        'HD02_VD01': 'margin_error'})

# Display the first few rows of the DataFrame to verify the changes
df.head()
```

```python
# Function to parse the principal city and state from the MSA name
# Note: The MSA names are mostly consistent except for special cases like Louisville and Honolulu
def parse_city_state(msa_name):
    # First, get the city name by splitting the MSA name at the comma and taking the first part
    city_name = msa_name.split(',')[0]
    # Further split the city name at hyphens and slashes to handle cases like Louisville/Jefferson County, KY
    city_name = city_name.split('-')[0]
    city_name = city_name.split('/')[0]  # Louisville/Jefferson County, KY
    # Remove any 'Urban ' prefix from the city name for cases like Urban Honolulu
    city_name = city_name.replace('Urban ', '')  # Urban Honolulu
    
    # Now, get the state name by splitting the MSA name at the comma and taking the second part
    state_name = msa_name.split(',')[1]
    # Remove leading and trailing whitespace and split at spaces to handle cases like "KY (Part)"
    state_name = state_name.strip().split(' ')[0]
    # Further split the state name at hyphens to handle any additional formatting
    state_name = state_name.split('-')[0]
    
    # Combine the city and state into a single string in the format "City, State"
    city_state = '{}, {}'.format(city_name, state_name)
    return city_state

# Apply the parse_city_state function to the 'msa_name' column to create a new 'city_state' column
df['city_state'] = df['msa_name'].map(parse_city_state)

# Display the first few rows of the DataFrame to verify the changes
df.head()
```

```python
import pandas as pd
import time
import requests
import json

# Function to geocode each city/state to latitude and longitude for mapping
def geocode(row):
    pause = 0.5  # Pause for 0.5 seconds between requests to avoid overwhelming the server
    address = row['city_state']  # Extract the city/state address from the DataFrame row
    print(row.name, end=' ')  # Print the row index for tracking progress
    
    # Pause for the specified duration before making each request to avoid hammering the server
    time.sleep(pause)  
    
    # Geocode API URL with placeholders for the address and the API key
    api_key = 'YOUR API KEY'  # Replace with your actual API key
    url = 'https://maps.googleapis.com/maps/api/geocode/json?address={}&key={}'  
    
    # Fill in the placeholders with the actual address and API key
    request = url.format(address, api_key)  
    
    # Send the HTTP request to the server and get the response
    response = requests.get(request)  
    
    # Convert the response JSON string into a dictionary
    data = json.loads(response.text)  
    
    # If Google was able to geolocate the address, extract latitude and longitude from the response
    if len(data['results']) > 0:  
        latitude = data['results'][0]['geometry']['location']['lat']
        longitude = data['results'][0]['geometry']['location']['lng']
        return pd.Series({'latitude': latitude, 'longitude': longitude})
    else:
        # If no results were returned, print the response data for debugging
        print(data)  

# # Apply the geocode function to each row in the DataFrame
# df[['latitude', 'longitude']] = df.apply(geocode, axis=1)

# # Display the first few rows of the DataFrame to verify the changes
# df.head()
```

```python
# Apply the geocode function to each row in the DataFrame
# This returns a DataFrame 'lat_long' with two columns: 'latitude' and 'longitude'
lat_long = df.apply(geocode, axis=1)

# Concatenate the 'lat_long' DataFrame with the original DataFrame 'df'
# The axis=1 parameter indicates that we are concatenating columns
df = pd.concat([df, lat_long], axis=1)

# Display the first few rows of the updated DataFrame to verify the changes
df.head()
```

```python
# All done - save the DataFrame to a CSV file
# The to_csv method is used to write the DataFrame to a CSV file named 'rent_latlong.csv'
# The index=False parameter ensures that the DataFrame index is not included as a column in the CSV file
df.to_csv('rent_latlong.csv', index=False)
```
